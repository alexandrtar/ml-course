```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import (
    Input, Embedding, LSTM, Dense, Dropout,
    Bidirectional, LayerNormalization,
    GlobalAveragePooling1D, MultiHeadAttention,
    SpatialDropout1D
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import (
    EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
)
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import warnings
warnings.filterwarnings('ignore')

# Инициализация NLTK
nltk.download('wordnet')
nltk.download('stopwords')
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# 1. Улучшенная загрузка и предобработка данных
def load_and_preprocess_data():
    train_df = pd.read_csv('train.csv')
    test_df = pd.read_csv('test.csv')

    def advanced_text_cleaner(text):
        text = str(text).lower()
        text = re.sub(r'\([^)]*\)', '', text)
        text = re.sub(r'[^\w\s]', ' ', text)
        text = re.sub(r'\d+', ' ', text)
        text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
        text = ' '.join([word for word in text.split() if word not in stop_words and len(word) > 2])
        text = re.sub(r'\s+', ' ', text).strip()
        return text

    train_df['text'] = train_df['text'].apply(advanced_text_cleaner)
    test_df['text'] = test_df['text'].apply(advanced_text_cleaner)

    plt.figure(figsize=(12, 8))
    sns.countplot(y='genre', data=train_df, order=train_df['genre'].value_counts().index)
    plt.title('Распределение жанров')
    plt.show()

    label_encoder = LabelEncoder()
    train_df['label'] = label_encoder.fit_transform(train_df['genre'])
    num_classes = len(label_encoder.classes_)

    X_train, X_val, y_train, y_val = train_test_split(
        train_df['text'].values,
        train_df['label'].values,
        test_size=0.2,
        random_state=42,
        stratify=train_df['label'].values
    )

    return X_train, X_val, y_train, y_val, test_df, label_encoder, num_classes

# 2. Упрощенная версия Transformer Block без training параметра
class TransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super().__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.layernorm1 = LayerNormalization(epsilon=1e-6)
        self.layernorm2 = LayerNormalization(epsilon=1e-6)
        self.dropout1 = Dropout(rate)
        self.dropout2 = Dropout(rate)

    def call(self, inputs):
        attn_output = self.att(inputs, inputs)
        attn_output = self.dropout1(attn_output)
        out1 = self.layernorm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output)
        return self.layernorm2(out1 + ffn_output)

# 3. Построение модели с упрощенным TransformerBlock
def build_hybrid_model(vocab_size=15000, max_len=80, num_classes=27):
    inputs = Input(shape=(max_len,))
    embedding = Embedding(vocab_size, 128, input_length=max_len, mask_zero=True)(inputs)

    # BiLSTM слой
    lstm_out = Bidirectional(LSTM(64, return_sequences=True))(embedding)

    # Transformer блок
    x = TransformerBlock(embed_dim=128, num_heads=4, ff_dim=64)(lstm_out)
    x = GlobalAveragePooling1D()(x)

    # Классификатор
    x = Dense(64, activation="relu")(x)
    x = Dropout(0.4)(x)
    outputs = Dense(num_classes, activation="softmax")(x)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(
        optimizer=Adam(learning_rate=3e-4, clipnorm=1.0),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

# 4. Подготовка данных
def prepare_data(X_train, X_val, vocab_size=15000, max_length=80):
    tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
    tokenizer.fit_on_texts(X_train)

    train_sequences = tokenizer.texts_to_sequences(X_train)
    val_sequences = tokenizer.texts_to_sequences(X_val)

    train_padded = pad_sequences(train_sequences, maxlen=max_length, padding='post', truncating='post')
    val_padded = pad_sequences(val_sequences, maxlen=max_length, padding='post', truncating='post')

    return train_padded, val_padded, tokenizer

# 5. Веса классов
def compute_class_weights(y_train):
    class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
    median_weight = np.median(class_weights)
    class_weights = np.where(class_weights > median_weight, class_weights * 1.5, class_weights * 2.0)
    return dict(enumerate(class_weights))

# 6. Callbacks
def get_callbacks():
    return [
        EarlyStopping(monitor='val_accuracy', patience=5, mode='max', restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6),
        ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_accuracy')
    ]

# 7. Оценка модели
def evaluate_model(model, X_val, y_val, label_encoder):
    y_pred = np.argmax(model.predict(X_val), axis=1)

    print("\nClassification Report:")
    print(classification_report(y_val, y_pred, target_names=label_encoder.classes_, zero_division=0))

    top_n = min(15, len(label_encoder.classes_))
    top_classes = label_encoder.classes_[:top_n]

    plt.figure(figsize=(14, 12))
    cm = confusion_matrix(y_val, y_pred, labels=range(top_n))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=top_classes, yticklabels=top_classes)
    plt.title(f'Confusion Matrix (Top {top_n} Classes)')
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    return y_pred

# 8. Сохранение предсказаний
def save_predictions(model, test_df, label_encoder, tokenizer, max_length=80):
    test_sequences = tokenizer.texts_to_sequences(test_df['text'])
    test_padded = pad_sequences(test_sequences, maxlen=max_length, padding='post', truncating='post')
    test_pred = np.argmax(model.predict(test_padded), axis=1)
    test_df['predicted_genre'] = label_encoder.inverse_transform(test_pred)
    test_df[['name', 'predicted_genre']].to_csv('predictions.csv', index=False)
    print("\nPredictions saved to predictions.csv")

# Основной блок
def main():
    np.random.seed(42)
    tf.random.set_seed(42)

    print("Загрузка и предобработка данных...")
    X_train, X_val, y_train, y_val, test_df, label_encoder, num_classes = load_and_preprocess_data()

    print("\nПодготовка данных для модели...")
    train_padded, val_padded, tokenizer = prepare_data(X_train, X_val)

    print("\nПостроение модели...")
    model = build_hybrid_model(vocab_size=15000, max_len=80, num_classes=num_classes)
    model.summary()

    print("\nВычисление весов классов...")
    class_weights = compute_class_weights(y_train)

    print("\nОбучение модели...")
    history = model.fit(
        train_padded, y_train,
        validation_data=(val_padded, y_val),
        epochs=30,
        batch_size=256,
        callbacks=get_callbacks(),
        class_weight=class_weights,
        verbose=1
    )

    # Визуализация обучения
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Validation')
    plt.title('Model Accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.legend()
    plt.show()

    print("\nОценка модели...")
    evaluate_model(model, val_padded, y_val, label_encoder)

    print("\nСохранение предсказаний...")
    save_predictions(model, test_df, label_encoder, tokenizer)

    print("\nОбучение завершено!")

if __name__ == "__main__":
    main()
```
