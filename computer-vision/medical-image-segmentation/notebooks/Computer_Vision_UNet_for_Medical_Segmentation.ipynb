```python
# ==================== –£–°–¢–ê–ù–û–í–ö–ê –í–°–ï–• –ó–ê–í–ò–°–ò–ú–û–°–¢–ï–ô ====================
print("üîß –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–µ–π...")
!pip install ultralytics opencv-python matplotlib scikit-learn tqdm pandas seaborn pycocotools --quiet

print("‚úÖ –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω—ã!")

# ==================== –ò–ú–ü–û–†–¢ –í–°–ï–• –ë–ò–ë–õ–ò–û–¢–ï–ö ====================
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torchvision
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from tqdm import tqdm
from ultralytics import YOLO
from sklearn.metrics import precision_score, recall_score, f1_score
import json
from PIL import Image

print("üìö –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")
print(f"üêç PyTorch version: {torch.__version__}")
print(f"üîÆ CUDA available: {torch.cuda.is_available()}")

# ==================== –ú–ï–î–ò–¶–ò–ù–°–ö–ê–Ø –°–ï–ì–ú–ï–ù–¢–ê–¶–ò–Ø (UNET) ====================

class DoubleConv(nn.Module):
    """(convolution => [BN] => ReLU) * 2"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.double_conv(x)

class UNet(nn.Module):
    def __init__(self, n_channels=1, n_classes=1):
        super(UNet, self).__init__()
        self.inc = DoubleConv(n_channels, 64)
        self.down1 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(64, 128))
        self.down2 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(128, 256))
        self.down3 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(256, 512))
        self.down4 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(512, 1024))

        self.up1 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.conv1 = DoubleConv(1024, 512)
        self.up2 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.conv2 = DoubleConv(512, 256)
        self.up3 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.conv3 = DoubleConv(256, 128)
        self.up4 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.conv4 = DoubleConv(128, 64)
        self.outc = nn.Conv2d(64, n_classes, kernel_size=1)

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)

        x = self.up1(x5)
        x = torch.cat([x, x4], dim=1)
        x = self.conv1(x)
        x = self.up2(x)
        x = torch.cat([x, x3], dim=1)
        x = self.conv2(x)
        x = self.up3(x)
        x = torch.cat([x, x2], dim=1)
        x = self.conv3(x)
        x = self.up4(x)
        x = torch.cat([x, x1], dim=1)
        x = self.conv4(x)
        return torch.sigmoid(self.outc(x))

def dice_coeff(pred, target):
    smooth = 1.
    pred_flat = pred.view(-1)
    target_flat = target.view(-1)
    intersection = (pred_flat * target_flat).sum()
    return (2. * intersection + smooth) / (pred_flat.sum() + target_flat.sum() + smooth)

def iou_score(pred, target):
    pred_bin = (pred > 0.5).float()
    target_bin = (target > 0.5).float()
    intersection = (pred_bin * target_bin).sum()
    union = pred_bin.sum() + target_bin.sum() - intersection
    return (intersection + 1e-6) / (union + 1e-6)

def create_medical_dataset():
    """–°–æ–∑–¥–∞–Ω–∏–µ —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω—ã—Ö —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –º–µ–¥–∏—Ü–∏–Ω—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö"""
    print("üîÑ –°–æ–∑–¥–∞–Ω–∏–µ —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –º–µ–¥–∏—Ü–∏–Ω—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö...")

    os.makedirs('/content/medical_data/train/images', exist_ok=True)
    os.makedirs('/content/medical_data/train/masks', exist_ok=True)
    os.makedirs('/content/medical_data/val/images', exist_ok=True)
    os.makedirs('/content/medical_data/val/masks', exist_ok=True)

    for split in ['train', 'val']:
        n_samples = 50 if split == 'train' else 10

        for i in range(n_samples):
            # –°–æ–∑–¥–∞–µ–º —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–æ–µ –º–µ–¥–∏—Ü–∏–Ω—Å–∫–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Å —à—É–º–æ–º
            img = np.random.normal(0.5, 0.2, (256, 256))

            # –î–æ–±–∞–≤–ª—è–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ "–æ—Ä–≥–∞–Ω–æ–≤" - —ç–ª–ª–∏–ø—Å—ã —Ä–∞–∑–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
            centers = np.random.randint(30, 226, (np.random.randint(2, 5), 2))
            radii_x = np.random.randint(15, 50, len(centers))
            radii_y = np.random.randint(20, 60, len(centers))

            for center, rx, ry in zip(centers, radii_x, radii_y):
                y, x = np.ogrid[:256, :256]
                # –≠–ª–ª–∏–ø—Å –≤–º–µ—Å—Ç–æ –∫—Ä—É–≥–∞ –¥–ª—è —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–æ—Å—Ç–∏
                mask_ellipse = ((x - center[0]) / rx)**2 + ((y - center[1]) / ry)**2 <= 1
                img[mask_ellipse] += np.random.normal(0.3, 0.1)

            # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Ç–µ–∫—Å—Ç—É—Ä—ã
            img = np.clip(img, 0, 1)
            img += np.random.normal(0, 0.05, (256, 256))  # —à—É–º

            # –°–æ–∑–¥–∞–µ–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –º–∞—Å–∫–∏
            mask = np.zeros((256, 256))
            for center, rx, ry in zip(centers, radii_x, radii_y):
                y, x = np.ogrid[:256, :256]
                mask_ellipse = ((x - center[0]) / (rx * 0.8))**2 + ((y - center[1]) / (ry * 0.8))**2 <= 1
                mask[mask_ellipse] = 1

            # –°–æ—Ö—Ä–∞–Ω—è–µ–º
            plt.imsave(f'/content/medical_data/{split}/images/sample_{i:03d}.png', img, cmap='gray')
            plt.imsave(f'/content/medical_data/{split}/masks/sample_{i:03d}.png', mask, cmap='gray')

    print("‚úÖ –°–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –º–µ–¥–∏—Ü–∏–Ω—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–∑–¥–∞–Ω—ã")

class MedicalDataset(Dataset):
    def __init__(self, data_dir, augment=False):
        self.data_dir = data_dir
        self.augment = augment
        self.image_dir = os.path.join(data_dir, 'images')
        self.mask_dir = os.path.join(data_dir, 'masks')
        self.image_files = sorted([f for f in os.listdir(self.image_dir) if f.endswith('.png')])

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_name = self.image_files[idx]
        img_path = os.path.join(self.image_dir, img_name)
        mask_path = os.path.join(self.mask_dir, img_name)

        # –ó–∞–≥—Ä—É–∑–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
        image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

        if image is None or mask is None:
            # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç—ã–µ –¥–∞–Ω–Ω—ã–µ –µ—Å–ª–∏ —Ñ–∞–π–ª—ã –Ω–µ –Ω–∞–π–¥–µ–Ω—ã
            image = np.random.rand(256, 256).astype(np.float32)
            mask = np.zeros((256, 256), dtype=np.float32)
        else:
            image = image.astype(np.float32) / 255.0
            mask = (mask > 0.5).astype(np.float32)

        # –ü—Ä–æ—Å—Ç—ã–µ –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏–∏ —Å –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏–µ–º –¥–ª—è –∏–∑–±–µ–∂–∞–Ω–∏—è –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö —à–∞–≥–æ–≤
        if self.augment and np.random.random() > 0.5:
            # –ì–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–µ –æ—Ç—Ä–∞–∂–µ–Ω–∏–µ
            image = np.fliplr(image).copy()  # .copy() –¥–ª—è –∏–∑–±–µ–∂–∞–Ω–∏—è –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö —à–∞–≥–æ–≤
            mask = np.fliplr(mask).copy()

        if self.augment and np.random.random() > 0.5:
            # –í–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–µ –æ—Ç—Ä–∞–∂–µ–Ω–∏–µ
            image = np.flipud(image).copy()
            mask = np.flipud(mask).copy()

        # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è
        image = (image - 0.5) / 0.5  # Normalize to [-1, 1]

        # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ —Ç–µ–Ω–∑–æ—Ä—ã —Å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é
        image_tensor = torch.from_numpy(image).unsqueeze(0).float()  # [1, 256, 256]
        mask_tensor = torch.from_numpy(mask).unsqueeze(0).float()    # [1, 256, 256]

        return image_tensor, mask_tensor

def train_medical_segmentation():
    """–û–±—É—á–µ–Ω–∏–µ –º–µ–¥–∏—Ü–∏–Ω—Å–∫–æ–π —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏–∏"""
    print("üè• –ó–∞–ø—É—Å–∫ –æ–±—É—á–µ–Ω–∏—è –º–µ–¥–∏—Ü–∏–Ω—Å–∫–æ–π —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏–∏...")

    # –°–æ–∑–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ –µ—Å–ª–∏ –∏—Ö –Ω–µ—Ç
    if not os.path.exists('/content/medical_data'):
        create_medical_dataset()

    # –î–∞—Ç–∞—Å–µ—Ç—ã
    train_dataset = MedicalDataset('/content/medical_data/train', augment=True)
    val_dataset = MedicalDataset('/content/medical_data/val', augment=False)

    train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)

    print(f"üìä Train samples: {len(train_dataset)}")
    print(f"üìä Val samples: {len(val_dataset)}")

    # –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö
    sample_image, sample_mask = train_dataset[0]
    print(f"üìê –†–∞–∑–º–µ—Ä –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è: {sample_image.shape}")
    print(f"üìê –†–∞–∑–º–µ—Ä –º–∞—Å–∫–∏: {sample_mask.shape}")

    # –ú–æ–¥–µ–ª—å –∏ –æ–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = UNet().to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
    criterion = nn.BCELoss()

    # –û–±—É—á–µ–Ω–∏–µ
    best_dice = 0
    history = {'train_loss': [], 'val_loss': [], 'dice': [], 'iou': []}

    for epoch in range(10):
        # Train
        model.train()
        train_loss = 0
        train_batches = 0

        progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/10')
        for images, masks in progress_bar:
            try:
                images, masks = images.to(device), masks.to(device)

                optimizer.zero_grad()
                outputs = model(images)

                loss = criterion(outputs, masks)
                loss.backward()
                optimizer.step()

                train_loss += loss.item()
                train_batches += 1

                progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})

            except Exception as e:
                print(f"‚ùå –û—à–∏–±–∫–∞ –≤ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–º –±–∞—Ç—á–µ: {e}")
                continue

        if train_batches == 0:
            print("‚ö†Ô∏è –ù–µ—Ç –≤–∞–ª–∏–¥–Ω—ã—Ö —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –±–∞—Ç—á–µ–π, –ø—Ä–æ–ø—É—Å–∫–∞–µ–º —ç–ø–æ—Ö—É")
            continue

        # Validation
        model.eval()
        val_loss = 0
        dice_total = 0
        iou_total = 0
        val_batches = 0

        with torch.no_grad():
            for images, masks in val_loader:
                try:
                    images, masks = images.to(device), masks.to(device)

                    outputs = model(images)

                    val_loss += criterion(outputs, masks).item()
                    dice_total += dice_coeff(outputs, masks).item()
                    iou_total += iou_score(outputs, masks).item()
                    val_batches += 1

                except Exception as e:
                    print(f"‚ùå –û—à–∏–±–∫–∞ –≤ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–º –±–∞—Ç—á–µ: {e}")
                    continue

        if val_batches == 0:
            print("‚ö†Ô∏è –ù–µ—Ç –≤–∞–ª–∏–¥–Ω—ã—Ö –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã—Ö –±–∞—Ç—á–µ–π")
            continue

        # –ú–µ—Ç—Ä–∏–∫–∏
        train_loss /= train_batches
        val_loss /= val_batches
        dice = dice_total / val_batches
        iou = iou_total / val_batches

        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['dice'].append(dice)
        history['iou'].append(iou)

        print(f'Epoch {epoch+1:02d} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Dice: {dice:.4f} | IoU: {iou:.4f}')

        if dice > best_dice:
            best_dice = dice
            torch.save(model.state_dict(), 'best_medical_model.pth')
            print(f"üíæ –°–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –ª—É—á—à–∞—è –º–æ–¥–µ–ª—å —Å Dice: {dice:.4f}")

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    if history['dice']:
        visualize_medical_results(model, val_loader, device, history)
    else:
        print("‚ùå –ù–µ—Ç –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏")

    return model, history

def visualize_medical_results(model, val_loader, device, history):
    """–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –º–µ–¥–∏—Ü–∏–Ω—Å–∫–æ–π —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏–∏"""
    model.eval()

    # –ì—Ä–∞—Ñ–∏–∫–∏ –æ–±—É—á–µ–Ω–∏—è
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Val Loss')
    plt.legend()
    plt.title('Training History')
    plt.grid(True)

    plt.subplot(1, 3, 2)
    plt.plot(history['dice'], label='Dice', color='green')
    plt.title('Dice Coefficient')
    plt.grid(True)

    plt.subplot(1, 3, 3)
    plt.plot(history['iou'], label='IoU', color='orange')
    plt.title('IoU Score')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # –ü—Ä–∏–º–µ—Ä—ã –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    with torch.no_grad():
        try:
            images, masks = next(iter(val_loader))
            images, masks = images.to(device), masks.to(device)
            preds = model(images[:4])

            print("\nüéØ –ü—Ä–∏–º–µ—Ä—ã –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π:")
            for i in range(min(4, len(images))):
                fig, axes = plt.subplots(1, 3, figsize=(12, 4))

                # –î–µ–Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
                img_display = (images[i].cpu().squeeze() * 0.5 + 0.5).numpy()

                # Input
                axes[0].imshow(img_display, cmap='gray')
                axes[0].set_title('Input Image')
                axes[0].axis('off')

                # Ground Truth
                axes[1].imshow(masks[i].cpu().squeeze(), cmap='gray')
                axes[1].set_title('Ground Truth')
                axes[1].axis('off')

                # Prediction
                pred_display = preds[i].cpu().squeeze()
                axes[2].imshow(pred_display, cmap='gray')
                current_dice = dice_coeff(preds[i], masks[i]).item()
                axes[2].set_title(f'Prediction\nDice: {current_dice:.3f}')
                axes[2].axis('off')

                plt.tight_layout()
                plt.show()
        except Exception as e:
            print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏: {e}")

# ==================== INSTANCE SEGMENTATION (YOLO) ====================

class YOLOSegmenter:
    def __init__(self, model_name='yolov8n-seg.pt'):
        """–ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è YOLO —Å–µ–≥–º–µ–Ω—Ç–∞—Ç–æ—Ä–∞"""
        print(f"üîÑ –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏ {model_name}...")
        self.model = YOLO(model_name)
        self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        print(f"‚úÖ –ú–æ–¥–µ–ª—å –∑–∞–≥—Ä—É–∂–µ–Ω–∞ –Ω–∞ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {self.device}")

    def segment_image(self, image, conf_threshold=0.3, target_size=(512, 512)):
        """–°–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è"""
        try:
            # –ö–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏—è –µ—Å–ª–∏ –Ω—É–∂–Ω–æ
            if isinstance(image, str):  # –ø—É—Ç—å –∫ —Ñ–∞–π–ª—É
                image = cv2.imread(image)
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            elif image.dtype == np.float32:
                image = (image * 255).astype(np.uint8)
            elif len(image.shape) == 2:  # grayscale to RGB
                image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

            # –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞
            image_resized = cv2.resize(image, target_size)

            # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ
            results = self.model(image_resized, conf=conf_threshold)

            # –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å–∫–∏
            mask = np.zeros(target_size[::-1], dtype=np.float32)

            for result in results:
                if result.masks is not None:
                    for m in result.masks:
                        m_data = m.data[0].cpu().numpy()
                        m_resized = cv2.resize(m_data.astype(np.uint8),
                                             target_size,
                                             interpolation=cv2.INTER_NEAREST)
                        mask = np.logical_or(mask, m_resized > 0.5)

            return mask.astype(np.float32)

        except Exception as e:
            print(f"‚ùå –û—à–∏–±–∫–∞ —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏–∏: {e}")
            return np.zeros(target_size[::-1], dtype=np.float32)

    def create_detection_images(self, num_samples=3):
        """–°–æ–∑–¥–∞–Ω–∏–µ —Ç–µ—Å—Ç–æ–≤—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π —Å —Ä–∞–∑–Ω—ã–º–∏ –æ–±—ä–µ–∫—Ç–∞–º–∏"""
        images = []

        # –°–ø–∏—Å–æ–∫ –æ–±—ä–µ–∫—Ç–æ–≤ –¥–ª—è –¥–µ—Ç–µ–∫—Ü–∏–∏
        objects = [
            {'name': 'person', 'color': (255, 200, 200), 'shape': 'human'},
            {'name': 'car', 'color': (200, 200, 255), 'shape': 'rectangle'},
            {'name': 'ball', 'color': (200, 255, 200), 'shape': 'circle'},
            {'name': 'dog', 'color': (255, 255, 200), 'shape': 'animal'}
        ]

        for i in range(num_samples):
            # –°–æ–∑–¥–∞–µ–º —Ä–∞–∑–Ω–æ–æ–±—Ä–∞–∑–Ω—ã–π —Ñ–æ–Ω
            bg_color = np.random.randint(100, 200, 3)
            img = np.full((512, 512, 3), bg_color, dtype=np.uint8)

            # –î–æ–±–∞–≤–ª—è–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –æ–±—ä–µ–∫—Ç–æ–≤
            for obj in objects[:2]:  # –î–æ–±–∞–≤–ª—è–µ–º 2 –æ–±—ä–µ–∫—Ç–∞ –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
                color = obj['color']
                center = (np.random.randint(100, 412), np.random.randint(100, 412))
                size = np.random.randint(40, 100)

                if obj['shape'] == 'human':
                    # –£–ø—Ä–æ—â–µ–Ω–Ω–∞—è —Ñ–∏–≥—É—Ä–∞ —á–µ–ª–æ–≤–µ–∫–∞
                    cv2.rectangle(img,
                                 (center[0]-30, center[1]-80),
                                 (center[0]+30, center[1]+40),
                                 color, -1)  # —Ç–µ–ª–æ
                    cv2.circle(img, center, 25, color, -1)  # –≥–æ–ª–æ–≤–∞
                    cv2.rectangle(img,
                                 (center[0]-50, center[1]-30),
                                 (center[0]-30, center[1]+20),
                                 color, -1)  # –ª–µ–≤–∞—è —Ä—É–∫–∞
                    cv2.rectangle(img,
                                 (center[0]+30, center[1]-30),
                                 (center[0]+50, center[1]+20),
                                 color, -1)  # –ø—Ä–∞–≤–∞—è —Ä—É–∫–∞

                elif obj['shape'] == 'rectangle':
                    # –ü—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ (–∞–≤—Ç–æ–º–æ–±–∏–ª—å)
                    cv2.rectangle(img,
                                 (center[0]-size, center[1]-size//2),
                                 (center[0]+size, center[1]+size//2),
                                 color, -1)

                elif obj['shape'] == 'circle':
                    # –ö—Ä—É–≥ (–º—è—á)
                    cv2.circle(img, center, size, color, -1)

                else:  # animal
                    # –£–ø—Ä–æ—â–µ–Ω–Ω–æ–µ –∂–∏–≤–æ—Ç–Ω–æ–µ
                    cv2.ellipse(img, center, (size, size//2), 0, 0, 360, color, -1)

            images.append(img)

        return images

def run_instance_segmentation():
    """–ó–∞–ø—É—Å–∫ instance segmentation"""
    print("üöÄ –ó–∞–ø—É—Å–∫ Instance Segmentation —Å YOLOv8...")

    try:
        segmenter = YOLOSegmenter()

        # –°–æ–∑–¥–∞–µ–º —Ç–µ—Å—Ç–æ–≤—ã–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
        test_images = segmenter.create_detection_images(num_samples=3)

        print("üéØ –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Ä–∞–±–æ—Ç—ã YOLOv8...")
        for i, image in enumerate(test_images):
            # –°–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–æ—Ä–æ–≥–∞–º–∏
            for conf in [0.25, 0.5]:
                mask = segmenter.segment_image(image, conf_threshold=conf)

                # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
                fig, axes = plt.subplots(1, 3, figsize=(15, 5))

                # –û—Ä–∏–≥–∏–Ω–∞–ª
                axes[0].imshow(image)
                axes[0].set_title(f'Original Image {i+1}')
                axes[0].axis('off')

                # –ú–∞—Å–∫–∞
                axes[1].imshow(mask, cmap='gray')
                axes[1].set_title(f'Segmentation Mask\nConf: {conf}')
                axes[1].axis('off')

                # –ù–∞–ª–æ–∂–µ–Ω–∏–µ
                axes[2].imshow(image)
                axes[2].imshow(mask, cmap='jet', alpha=0.5)
                axes[2].set_title('Overlay')
                axes[2].axis('off')

                plt.tight_layout()
                plt.show()

        print("‚úÖ Instance segmentation –∑–∞–≤–µ—Ä—à–µ–Ω–∞ —É—Å–ø–µ—à–Ω–æ")
        return segmenter

    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –≤ instance segmentation: {e}")
        return None

# ==================== –ì–õ–ê–í–ù–´–ô –ó–ê–ü–£–°–ö ====================

def main():
    """–ì–ª–∞–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –∑–∞–ø—É—Å–∫–∞ –≤—Å–µ—Ö –ø—Ä–æ–µ–∫—Ç–æ–≤"""
    print("üéØ –ó–ê–ü–£–°–ö –ü–û–õ–ù–û–ì–û –ü–ê–ô–ü–õ–ê–ô–ù–ê –ö–û–ú–ü–¨–Æ–¢–ï–†–ù–û–ì–û –ó–†–ï–ù–ò–Ø")
    print("=" * 60)

    results = {}

    try:
        # 1. –ú–µ–¥–∏—Ü–∏–Ω—Å–∫–∞—è —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è
        print("\n" + "="*50)
        print("1. üè• –ú–ï–î–ò–¶–ò–ù–°–ö–ê–Ø –°–ï–ì–ú–ï–ù–¢–ê–¶–ò–Ø (UNet)")
        print("="*50)
        medical_model, medical_history = train_medical_segmentation()
        if medical_history and medical_history['dice']:
            results['medical'] = {
                'model': medical_model,
                'history': medical_history,
                'best_dice': max(medical_history['dice'])
            }
        else:
            results['medical'] = None
            print("‚ùå –ú–µ–¥–∏—Ü–∏–Ω—Å–∫–∞—è —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è –Ω–µ –¥–∞–ª–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤")
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –≤ –º–µ–¥–∏—Ü–∏–Ω—Å–∫–æ–π —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏–∏: {e}")
        import traceback
        traceback.print_exc()
        results['medical'] = None

    try:
        # 2. Instance segmentation
        print("\n" + "="*50)
        print("2. üë§ INSTANCE SEGMENTATION (YOLOv8)")
        print("="*50)
        segmenter = run_instance_segmentation()
        results['instance'] = {
            'segmenter': segmenter
        }
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –≤ instance segmentation: {e}")
        results['instance'] = None

    # 3. –ò—Ç–æ–≥–æ–≤—ã–π –æ—Ç—á–µ—Ç
    print("\n" + "="*50)
    print("3. üìä –ò–¢–û–ì–û–í–´–ô –û–¢–ß–ï–¢")
    print("="*50)

    print("üéâ –ü–ê–ô–ü–õ–ê–ô–ù –ó–ê–í–ï–†–®–ï–ù!")
    print("\nüìà –†–ï–ó–£–õ–¨–¢–ê–¢–´:")

    if results.get('medical') and results['medical']:
        best_dice = results['medical']['best_dice']
        print(f"üè• Medical Segmentation:")
        print(f"   ‚Ä¢ –õ—É—á—à–∏–π Dice: {best_dice:.4f}")
        if results['medical']['history'] and results['medical']['history']['iou']:
            best_iou = max(results['medical']['history']['iou'])
            print(f"   ‚Ä¢ –õ—É—á—à–∏–π IoU: {best_iou:.4f}")
    else:
        print(f"üè• Medical Segmentation: –ù–µ –∑–∞–≤–µ—Ä—à–µ–Ω–∞")

    if results.get('instance') and results['instance']['segmenter']:
        print(f"üë§ Instance Segmentation:")
        print(f"   ‚Ä¢ YOLOv8 —É—Å–ø–µ—à–Ω–æ –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω")
        print(f"   ‚Ä¢ –ú–æ–¥–µ–ª—å: yolov8n-seg.pt")
    else:
        print(f"üë§ Instance Segmentation: –ù–µ –∑–∞–≤–µ—Ä—à–µ–Ω–∞")

    print("\nüìÅ –°–û–ó–î–ê–ù–ù–´–ï –§–ê–ô–õ–´:")
    created_files = []
    if os.path.exists('best_medical_model.pth'):
        created_files.append('best_medical_model.pth')
    if os.path.exists('/content/medical_data'):
        created_files.append('/content/medical_data/ (—Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ)')

    for file in created_files:
        print(f"   ‚Ä¢ {file}")

    if not created_files:
        print("   ‚Ä¢ –ù–µ—Ç —Å–æ–∑–¥–∞–Ω–Ω—ã—Ö —Ñ–∞–π–ª–æ–≤")

    return results

# ==================== –ó–ê–ü–£–°–ö ====================

if __name__ == "__main__":
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–æ—Å—Ç—É–ø–Ω–æ—Å—Ç–∏ GPU
    if torch.cuda.is_available():
        print(f"üéØ GPU –¥–æ—Å—Ç—É–ø–µ–Ω: {torch.cuda.get_device_name(0)}")
    else:
        print("‚ö†Ô∏è GPU –Ω–µ –¥–æ—Å—Ç—É–ø–µ–Ω, –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è CPU")

    # –ó–∞–ø—É—Å–∫ –≥–ª–∞–≤–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
    try:
        final_results = main()

        print("\n" + "üéØ" * 30)
        print("üöÄ –ü–ê–ô–ü–õ–ê–ô–ù –ó–ê–í–ï–†–®–ï–ù!")
        print("üéØ" * 30)
    except Exception as e:
        print(f"‚ùå –ö—Ä–∏—Ç–∏—á–µ—Å–∫–∞—è –æ—à–∏–±–∫–∞: {e}")
        import traceback
        traceback.print_exc()
```
